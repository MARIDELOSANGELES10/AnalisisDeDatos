```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
import pandas as pd

# Leer el archivo de datos (Asegúrese de que la ruta sea la correcta)
df = pd.read_csv('C:/anaconda_data/data.csv')

# Mostrar el dataframe
print(df.head())

# Definir variables dependiente e independiente
independent_var = df[['metro']]
dependent_var = df['precio']

# Añadir una constante al modelo (intercepto)
independent_var = sm.add_constant(independent_var)

# Crear un modelo OLS (Ordinary Least Squares)
model = sm.OLS(dependent_var, independent_var).fit()

# Hacer predicciones
predicted_values = model.predict(independent_var)

# Mostrar parámetros del modelo
print("Intercepto: ", model.params[0])
print("Pendiente: ", model.params[1])

# Visualizar los resultados
plt.scatter(df['metro'], df['precio'], label='Datos Observados')
plt.plot(df['metro'], predicted_values, color='red', label='Modelo de Regresión')
plt.xlabel('Metros cuadrados')
plt.ylabel('Precio')
plt.legend()
plt.show()
```
